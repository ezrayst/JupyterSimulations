```python
# Import Matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

## Traction Controlled

```python
# Getting more data

stress_iteration_1 = np.genfromtxt("master/stress_0.txt")
stress_iteration_2 = np.genfromtxt("iteration/stress_0.txt")

strain_iteration_1 = np.genfromtxt("master/strain_0.txt")
strain_iteration_2 = np.genfromtxt("iteration/strain_0.txt")

acceleration_iteration_1 = np.genfromtxt("master/acceleration_0.txt")
acceleration_iteration_2 = np.genfromtxt("iteration/acceleration_0.txt")

velocity_iteration_1 = np.genfromtxt("master/velocity_0.txt")
velocity_iteration_2 = np.genfromtxt("iteration/velocity_0.txt")
```

```python
# Plot Comparison

plt.figure(figsize=(16,48))

plt.subplot(811)
plt.plot(strain_iteration_1[:,0], strain_iteration_1[:,1] * 100, 'b', label='No Iteration')
plt.plot(strain_iteration_2[:,0], strain_iteration_2[:,1] * 100, 'r--', label='With Iteration')
plt.axis([0, 100000, 0, 0.0006])
plt.legend(loc='lower right')
plt.title('Strain - X')
plt.xlabel('Time Step [10^-5 s]')
plt.ylabel('Strain [%]')
plt.grid(True)

plt.subplot(812)
plt.plot(strain_iteration_1[:,0], strain_iteration_1[:,2] * 100, 'b', label='No Iteration')
plt.plot(strain_iteration_2[:,0], strain_iteration_2[:,2] * 100, 'r--', label='With Iteration')
plt.axis([0, 100000, -0.003, 0 ])
plt.legend(loc='upper right')
plt.title('Strain - Y')
plt.xlabel('Time Step [10^-5 s]')
plt.ylabel('Strain [%]')
plt.grid(True)

plt.subplot(813)
plt.plot(stress_iteration_1[:,0], stress_iteration_1[:,1], 'b', label='No Iteration')
plt.plot(stress_iteration_2[:,0], stress_iteration_2[:,1], 'r--', label='With Iteration')
plt.axis([0, 100000, -20, 5])
plt.legend(loc='lower right')
plt.title('Stress - X')
plt.xlabel('Time Step [10^-5 s]')
plt.ylabel('Stress [Pa]')
plt.grid(True)

plt.subplot(814)
plt.plot(stress_iteration_1[:,0], stress_iteration_1[:,2], 'b', label='No Iteration')
plt.plot(stress_iteration_2[:,0], stress_iteration_2[:,2], 'r--', label='With Iteration')
plt.axis([0, 100000, -300, 0])
plt.legend(loc='upper right')
plt.title('Stress - Y')
plt.xlabel('Time Step [10^-5 s]')
plt.ylabel('Stress [Pa]')
plt.grid(True)

plt.subplot(815)
plt.plot(acceleration_iteration_1[:,0], acceleration_iteration_1[:,1], 'b', label='No Iteration')
plt.plot(acceleration_iteration_2[:,0], acceleration_iteration_2[:,1], 'r--', label='With Iteration')
plt.axis([0, 100000, -0.001, 0.2])
plt.legend(loc='lower right')
plt.title('Acceleration - X')
plt.xlabel('Time Step [10^-5 s]')
plt.ylabel('Acceleration [m / s^2]')
plt.grid(True)

plt.subplot(816)
plt.plot(acceleration_iteration_1[:,0], acceleration_iteration_1[:,2], 'b', label='No Iteration')
plt.plot(acceleration_iteration_2[:,0], acceleration_iteration_2[:,2], 'r--', label='With Iteration')
plt.axis([0, 100000, -0.4, 0.001])
plt.legend(loc='lower right')
plt.title('Acceleration - Y')
plt.xlabel('Time Step [10^-5 s]')
plt.ylabel('Acceleration [m / s^2]')
plt.grid(True)

plt.subplot(817)
plt.plot(velocity_iteration_1[:,0], velocity_iteration_1[:,1], 'b', label='No Iteration')
plt.plot(velocity_iteration_2[:,0], velocity_iteration_2[:,1], 'r--', label='With Iteration')
plt.axis([0, 100000, -0.000001, 0.00000125])
plt.legend(loc='upper right')
plt.title('Velocity - X')
plt.xlabel('Time Step [10^-5 s]')
plt.ylabel('Velocity [m / s]')
plt.grid(True)

plt.subplot(818)
plt.plot(velocity_iteration_1[:,0], velocity_iteration_1[:,2], 'b', label='No Iteration')
plt.plot(velocity_iteration_2[:,0], velocity_iteration_2[:,2], 'r--', label='With Iteration')
plt.axis([0, 100000, -0.00001, 0.00001])
plt.legend(loc='lower right')
plt.title('Velocity - Y')
plt.xlabel('Time Step [10^-5 s]')
plt.ylabel('Velocity [m / s]')
plt.grid(True)

plt.show()
```
