```python
# Plot Comparison

plt.figure(figsize=(16,18))

plt.subplot(311)
plt.plot(noinitialstress_stress_20[:,0], noinitialstress_stress_20[:,1], 'r', label='No Initial Stress')
plt.plot(initialstress_stress_20[:,0], initialstress_stress_20[:,1] , 'b', label='With Initial Stress')
plt.plot(damping02_stress_20[:,0], damping02_stress_20[:,1], 'g', label='2% Damping')
plt.plot(damping05_stress_20[:,0], damping05_stress_20[:,1], 'y', label='5% Damping')
plt.axis([0, 18000, -1000, 0])
plt.legend(loc='upper right')
plt.title('Stress - Y @ Point 20 (Bottom Center)')
plt.xlabel('Time Step [10^-5 s]')
plt.ylabel('Stress [Pa]')
plt.grid(True)

plt.subplot(312)
plt.plot(noinitialstress_stress_1[:,0], noinitialstress_stress_1[:,1], 'r', label='No Initial Stress')
plt.plot(initialstress_stress_1[:,0], initialstress_stress_1[:,1] , 'b', label='With Initial Stress')
plt.plot(damping02_stress_1[:,0], damping02_stress_1[:,1], 'g', label='2% Damping')
plt.plot(damping05_stress_1[:,0], damping05_stress_1[:,1], 'y', label='5% Damping')
plt.axis([0, 18000, -4000, 1000])
plt.legend(loc='upper right')
plt.title('Stress - Y @ Point 1 (Bottom Left)')
plt.xlabel('Time Step [10^-5 s]')
plt.ylabel('Stress [Pa]')
plt.grid(True)

plt.subplot(313)
plt.plot(noinitialstress_stress_420[:,0], noinitialstress_stress_420[:,1], 'r', label='No Initial Stress')
plt.plot(initialstress_stress_420[:,0], initialstress_stress_420[:,1] , 'b', label='With Initial Stress')
plt.plot(damping02_stress_420[:,0], damping02_stress_420[:,1], 'g', label='2% Damping')
plt.plot(damping05_stress_420[:,0], damping05_stress_420[:,1], 'y', label='5% Damping')
plt.axis([0, 18000, -1000, 0])
plt.legend(loc='lower right')
plt.title('Stress - Y @ Point 420 (Middle Center)')
plt.xlabel('Time Step [10^-5 s]')
plt.ylabel('Stress [Pa]')
plt.grid(True)


plt.show()
```
