```python
# Import Matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

## 4 Velocity on top and bottom

```python
# Getting more data

stress_iteration_1 = np.genfromtxt("1_elem/stress_0.txt")
stress_iteration_2 = np.genfromtxt("4_elem/stress_0.txt")

strain_iteration_1 = np.genfromtxt("1_elem/strain_0.txt")
strain_iteration_2 = np.genfromtxt("4_elem/strain_0.txt")

acceleration_iteration_1 = np.genfromtxt("1_elem/acceleration_0.txt")
acceleration_iteration_2 = np.genfromtxt("4_elem/acceleration_0.txt")

velocity_iteration_1 = np.genfromtxt("1_elem/velocity_0.txt")
velocity_iteration_2 = np.genfromtxt("4_elem/velocity_0.txt")
```

```python
# Plot Comparison

plt.figure(figsize=(16,48))

plt.subplot(811)
plt.plot(strain_iteration_1[:,0], strain_iteration_1[:,1] * 100, 'b', label='1 Element')
plt.plot(strain_iteration_2[:,0], strain_iteration_2[:,1] * 100, 'r--', label='4 Elements')
plt.axis([0, 100000, 0, 0.5])
plt.legend(loc='upper right')
plt.title('Strain - X')
plt.xlabel('Time Step [10^-5 s]')
plt.ylabel('Strain [%]')
plt.grid(True)

plt.subplot(812)
plt.plot(strain_iteration_1[:,0], strain_iteration_1[:,2] * 100, 'b', label='1 Element')
plt.plot(strain_iteration_2[:,0], strain_iteration_2[:,2] * 100, 'r--', label='4 Elements')
plt.axis([0, 100000, -1, 0 ])
plt.legend(loc='upper right')
plt.title('Strain - Y')
plt.xlabel('Time Step [10^-5 s]')
plt.ylabel('Strain [%]')
plt.grid(True)

plt.subplot(813)
plt.plot(stress_iteration_1[:,0], stress_iteration_1[:,1], 'b', label='1 Element')
plt.plot(stress_iteration_2[:,0], stress_iteration_2[:,1], 'r--', label='4 Elements')
plt.axis([0, 100000, -10, 5])
plt.legend(loc='lower right')
plt.title('Stress - X')
plt.xlabel('Time Step [10^-5 s]')
plt.ylabel('Stress [Pa]')
plt.grid(True)

plt.subplot(814)
plt.plot(stress_iteration_1[:,0], stress_iteration_1[:,2], 'b', label='1 Element')
plt.plot(stress_iteration_2[:,0], stress_iteration_2[:,2], 'r--', label='4 Elements')
plt.axis([0, 100000, -100000, 0])
plt.legend(loc='upper right')
plt.title('Stress - Y')
plt.xlabel('Time Step [10^-5 s]')
plt.ylabel('Stress [Pa]')
plt.grid(True)

plt.subplot(815)
plt.plot(acceleration_iteration_1[:,0], acceleration_iteration_1[:,1], 'b', label='1 Element')
plt.plot(acceleration_iteration_2[:,0], acceleration_iteration_2[:,1], 'r--', label='4 Elements')
plt.axis([0, 100000, -0.2, 1.2])
plt.legend(loc='lower right')
plt.title('Acceleration - X')
plt.xlabel('Time Step [10^-5 s]')
plt.ylabel('Acceleration [m / s^2]')
plt.grid(True)

plt.subplot(816)
plt.plot(acceleration_iteration_1[:,0], acceleration_iteration_1[:,2], 'b', label='1 Element')
plt.plot(acceleration_iteration_2[:,0], acceleration_iteration_2[:,2], 'r--', label='4 Elements')
plt.axis([0, 100000, -0.001, 0.001])
plt.legend(loc='lower right')
plt.title('Acceleration - Y')
plt.xlabel('Time Step [10^-5 s]')
plt.ylabel('Acceleration [m / s^2]')
plt.grid(True)

plt.subplot(817)
plt.plot(velocity_iteration_1[:,0], velocity_iteration_1[:,1], 'b', label='1 Element')
plt.plot(velocity_iteration_2[:,0], velocity_iteration_2[:,1], 'r--', label='4 Elements')
plt.axis([0, 100000, -0.0001, 0.0001])
plt.legend(loc='upper left')
plt.title('Velocity - X')
plt.xlabel('Time Step [10^-5 s]')
plt.ylabel('Velocity [m / s]')
plt.grid(True)

plt.subplot(818)
plt.plot(velocity_iteration_1[:,0], velocity_iteration_1[:,2], 'b', label='1 Element')
plt.plot(velocity_iteration_2[:,0], velocity_iteration_2[:,2], 'r--', label='4 Elements')
plt.axis([0, 100000, -0.0003, 0.0003])
plt.legend(loc='lower right')
plt.title('Velocity - Y')
plt.xlabel('Time Step [10^-5 s]')
plt.ylabel('Velocity [m / s]')
plt.grid(True)

plt.show()
```
