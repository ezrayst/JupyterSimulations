# MPM Damping for Two Phase

- Element test with left boundary horizontally restrained, and bottom boundary vertically restrained.
- 4 Material points equally spaced in the element.
- Tractions of -10,000 are applied to all 4 material points.
- Initial stresses and pressure are 0.

```python
# Import Matplotlib
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Image
```

```python
# Getting more data

nodamping = np.genfromtxt("nodamping/stress_0.txt")
withdamping = np.genfromtxt("damping/stress_0.txt")
analytical = np.genfromtxt("analytical.txt")
```

```python
# Plot Comparison

plt.figure(figsize=(16,6))

plt.plot(analytical[:,0], analytical[:,2], 'k--', label="Analytical")
plt.plot(nodamping[:,0], nodamping[:,2], 'r', label='No Damping')
plt.plot(withdamping[:,0], withdamping[:,2] , 'b', label='5% Damping')
plt.axis([0, 20, -12000, 0])
plt.legend(loc='upper right')
plt.title('Stress - Y')
plt.xlabel('Time Step [10^-7 s]')
plt.ylabel('Stress [Pa]')
plt.grid(True)


plt.show()
```
